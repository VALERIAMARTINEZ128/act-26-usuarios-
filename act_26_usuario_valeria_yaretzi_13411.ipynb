```python
# 1. Definimos la Clase Usuario 🏗️
class Usuario:
    def __init__(self, nombre, contrasena, rol):
        # Atributos: nombre, contraseña, rol 📌
        self.nombre = nombre
        self.contrasena = contrasena
        self.rol = rol

    # Función para leer los datos del teclado ⌨️
    def leerdatos(self):
        self.nombre = input("Ingrese el nombre de usuario: ")
        self.contrasena = input("Ingrese la contraseña: ")
        self.rol = input("Ingrese el rol (Admin/Invitado): ")
        return self.nombre, self.contrasena, self.rol

    # Función para mostrar los datos en pantalla 🖥️
    def mostrardatos(self):
        print("Nombre: ", self.nombre)
        print("Contraseña: ", self.contrasena)
        print("Rol: ", self.rol)
        return self.nombre, self.contrasena, self.rol
# Fin de la clase Usuario
# --- 2. Creación de Objetos (Instancias) 👥 ---
# Objeto 1
usuario1 = Usuario("", "", "")
print("--- DATOS USUARIO 1 ---")
usuario1.leerdatos()
usuario1.mostrardatos()
print("*-*-*-*")

# Objeto 2
usuario2 = Usuario("", "", "")
print("--- DATOS USUARIO 2 ---")
usuario2.leerdatos()
usuario2.mostrardatos()
print("*-*-*-*")

# Pie de página
print("software realizado por valeria yaretzi lopez martinez NL = 32")
```
